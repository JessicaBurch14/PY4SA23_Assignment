##### Task 3 - function to calculate the distance between two coordinates using the haversine formula

Haversine formula: 
a = sin²(φB - φA/2) + cos φA * cos φB * sin²(λB - λA/2)

c = 2 * atan2( √a, √(1−a) )

d = R ⋅ c

where:

φ = latitude 

λ = longitude 

R = earth's radius (mean radius = 6,371km)

These are my two chosen coordinates: 

A:(52.408054, -1.510556)- Coventry 

B:(56.3387, -2.7990) - St Andrews 

First I am converting these coordinate angles into radians using the math library so that the formula produces the correct result. Then I am putting these radian coordinates into the first part of the formula (a) and then printing the result. Then I am substituting this result into the second part of the formula (c) and printing the result. Next I add this result into the last part of thr formula (d) along with the value of R and printing the final result 

```python
## calculating the first part of the formula (a) 

# a = sin²(φB - φA/2) + cos φA * cos φB * sin²(λB - λA/2)
# φB = Blat , φA = Alat , λB = Blong , λA = Along 

# converting angles into radians
Alat= math.radians(52.408054)
Blat= math.radians(56.3387)
Along= math.radians(-1.510556)
Blong= math.radians(-2.7990)

# putting the coordinates into the formula 
# φB - φA/2 = Clat
# λB - λA/2 = Clong 
Clat= Blat-Alat
Clong= Blong-Along
a_answer= (math.sin(Clat/2)**2) + (math.cos(Alat)) * (math.cos(Blat)) * (math.sin(Clong/2)**2)
print(a_answer)
# a=0.001218867435140178

## calculating the second part of the formula (c)

# c = 2 * atan2( √a, √(1−a) )

# putting a into the c formula 
c_answer= 2 * math.atan2(math.sqrt(a_answer), math.sqrt(1-a_answer))
print(c_answer)
# c=0.06983875628674953

## calculating the last part of the formula (d)

# d = R ⋅ c
# R=6371

# putting c into the d formula 
d_answer= 6371 * c_answer 
print(d_answer)
# d= 444.94271630288125

#printing the result
print("The distance between Coventry and St Andrews is" + " " + str(d_answer)+ "km")
```
